### Tarjeta de débito

```python
# Clase: Tarjeta_Debito

class Tarjeta_Debito:

    def __init__(self, numero, saldo, banco, limite, estado):
        self.NumeroTarjeta = numero
        self.SaldoDisponible = saldo
        self.BancoEmisor = banco
        self.LimiteRetiro = limite
        self.EstadoTarjeta = estado

    def agregar_saldo(self):
        cantidad = float(input("Cantidad a depositar: $"))
        self.SaldoDisponible += cantidad
        print("Depósito realizado con éxito.")

    def retirar_dinero(self):
        if self.EstadoTarjeta == "Bloqueada":
            print("La tarjeta está bloqueada.")
        else:
            cantidad = float(input("Cantidad a retirar: $"))

            if cantidad <= self.SaldoDisponible and cantidad <= self.LimiteRetiro:
                self.SaldoDisponible -= cantidad
                print("Retiro exitoso.")
            else:
                print("Operación no permitida.")

    def consultar_saldo(self):
        print("Saldo disponible: $", self.SaldoDisponible)

    def bloquear_tarjeta(self):
        self.EstadoTarjeta = "Bloqueada"
        print("Tarjeta bloqueada correctamente.")

    def mostrar_datos(self):
        print("\n" + "="*60)
        print("{:<20} {:<35}".format("ATRIBUTO", "VALOR"))
        print("="*60)
        print("{:<20} {:<35}".format("Número", self.NumeroTarjeta))
        print("{:<20} {:<35}".format("Saldo", "$" + str(self.SaldoDisponible)))
        print("{:<20} {:<35}".format("Banco", self.BancoEmisor))
        print("{:<20} {:<35}".format("Límite", "$" + str(self.LimiteRetiro)))
        print("{:<20} {:<35}".format("Estado", self.EstadoTarjeta))
        print("="*60)


# Datos iniciales
numero = input("Número de tarjeta: ")
saldo = float(input("Saldo inicial: "))
banco = input("Banco emisor: ")
limite = float(input("Límite de retiro: "))
estado = "Activa"

tarjeta = Tarjeta_Debito(numero, saldo, banco, limite, estado)

while True:
    print("\n--- MENÚ TARJETA DE DÉBITO IA ---")
    print("1. Agregar saldo")
    print("2. Retirar dinero")
    print("3. Consultar saldo")
    print("4. Bloquear tarjeta")
    print("5. Mostrar datos")
    print("6. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        tarjeta.agregar_saldo()

    elif opcion == "2":
        tarjeta.retirar_dinero()

    elif opcion == "3":
        tarjeta.consultar_saldo()

    elif opcion == "4":
        tarjeta.bloquear_tarjeta()

    elif opcion == "5":
        tarjeta.mostrar_datos()

    elif opcion == "6":
        print("Programa finalizado.")
        break

    else:
        print("Opción inválida.")
```
